## Import matplotlib library

> **Mat**hematical **Plot**ting **Lib**rary

Matplotlib is a Python library that contains functions to create Data Visualizations

```python
import matplotlib.pyplot as plt
```

### Scatter plot (points)

```python
plt.scatter(x=groceries, y=prices)
```

### Bar plot

```python
plt.scatter(x=groceries, y=prices,s=50)
```

### Many others

https://matplotlib.org/stable/plot_types/index

## Customize the plot

### Change the size

```python
plt.scatter(x=groceries, y=prices,s=50)
```

### Change the color

```python
plt.scatter(x=groceries, y=prices,c='yellow')
```

### Change the marker

https://matplotlib.org/stable/api/markers_api.html

```python
    plt.scatter(x=groceries, y=prices,s=50,c='red',marker='+')
```

### Change the border

```python
plt.scatter(x=groceries, y=prices,s=100,c='red',marker='X',edgecolors='black')
```
